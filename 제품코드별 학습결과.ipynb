```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./train.csv")
data
```

```python
data_A = data[(data['PRODUCT_CODE']=='A_31')]
data_O = data[(data['PRODUCT_CODE']=='O_31')]
data_T = data[(data['PRODUCT_CODE']=='T_31')]
```

```python
def nan_check(data):
  total=data.isnull().sum().sort_values(ascending=False)
  percent_1 = data.isnull().sum()/data.isnull().count()*100
  percent_2 = (np.round(percent_1, 1)).sort_values(ascending=False)
  missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
  return missing_data

x_A=nan_check(data_A)
c_A1=x_A['%'] == 100
li_A=x_A[c_A1].index.to_list()

x_O=nan_check(data_O)
c_O1=x_O['%'] == 100
li_O=x_O[c_O1].index.to_list()

x_T=nan_check(data_T)
c_T1=x_T['%'] == 100
li_T=x_T[c_T1].index.to_list()
```

```python
for i in li_A:
  try:
    data_A.drop(i, axis=1, inplace=True)
  except:
    continue

for i in li_O:
  try:
    data_O.drop(i, axis=1, inplace=True)
  except:
    continue

for i in li_T:
  try:
    data_T.drop(i, axis=1, inplace=True)
  except:
    continue
```

```python
data_A.drop('TIMESTAMP', axis=1, inplace=True)
data_O.drop('TIMESTAMP', axis=1, inplace=True)
data_T.drop('TIMESTAMP', axis=1, inplace=True)
```

```python
trainYA=data_A.iloc[:,1]
trainYO=data_O.iloc[:,1]
trainYT=data_T.iloc[:,1]
```

```python
new_dataA=data_A.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})
new_dataO=data_O.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})
new_dataT=data_T.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})
```

```python
new_dataA
```

```python
trainXA=new_dataA.iloc[:, 3:]
trainXO=new_dataO.iloc[:, 3:]
trainXT=new_dataT.iloc[:, 3:]
```

```python
trainXT.drop('PRODUCT_CODE',axis=1,inplace=True)
```

```python
trainXA.drop('PRODUCT_CODE',axis=1,inplace=True)
trainXO.drop('PRODUCT_CODE',axis=1,inplace=True)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
param_xgb = {"max_depth": [10,30,50],
              "min_child_weight" : [1,3,6,10],
              "n_estimators": [200,300,500,1000]
              }    
```

```python
gscv_xgbA = GridSearchCV (estimator = xgb, param_grid = param_xgb, scoring ='accuracy', cv = 3, refit=True, n_jobs=1, verbose=2)
gscv_xgbA.fit(trainXA, trainYA)

gscv_xgbO = GridSearchCV (estimator = xgb, param_grid = param_xgb, scoring ='accuracy', cv = 3, refit=True, n_jobs=1, verbose=2)
gscv_xgbO.fit(trainXO, trainYO)

gscv_xgbT = GridSearchCV (estimator = xgb, param_grid = param_xgb, scoring ='accuracy', cv = 3, refit=True, n_jobs=1, verbose=2)
gscv_xgbT.fit(trainXT, trainYT)
```

```python
print('A_31의 XGB 파라미터: ', gscv_xgbA.best_params_)
print('A_31의 XGB 예측 정확도: {:.4f}'.format(gscv_xgbA.best_score_))

print('O_31의 XGB 파라미터: ', gscv_xgbO.best_params_)
print('O_31의 XGB 예측 정확도: {:.4f}'.format(gscv_xgbO.best_score_))

print('T_31의 XGB 파라미터: ', gscv_xgbT.best_params_)
print('T_31의 XGB 예측 정확도: {:.4f}'.format(gscv_xgbT.best_score_))
```

```python
data_test = pd.read_csv("./test.csv")
test_A = data_test[(data_test['PRODUCT_CODE']=='A_31')]
test_O = data_test[(data_test['PRODUCT_CODE']=='O_31')]
test_T = data_test[(data_test['PRODUCT_CODE']=='T_31')]

for i in li_A:
  try:
    test_A.drop(i, axis=1, inplace=True)
  except:
    continue

for i in li_O:
  try:
    test_O.drop(i, axis=1, inplace=True)
  except:
    continue

for i in li_T:
  try:
    test_T.drop(i, axis=1, inplace=True)
  except:
    continue

test_A.drop(['TIMESTAMP','PRODUCT_CODE'], axis=1, inplace=True)
test_O.drop(['TIMESTAMP','PRODUCT_CODE'], axis=1, inplace=True)
test_T.drop(['TIMESTAMP','PRODUCT_CODE'], axis=1, inplace=True)

new_testA=test_A.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})
new_testO=test_O.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})
new_testT=test_T.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})

testXA_num=new_testA.iloc[:,0]
testXO_num=new_testO.iloc[:,0]
testXT_num=new_testT.iloc[:,0]

testXA=new_testA.iloc[:,1:]
testXO=new_testO.iloc[:,1:]
testXT=new_testT.iloc[:,1:]
```

```python
testXA_num
```

```python
testA_num=testXA_num.reset_index()
testA_num.drop('index', axis=1, inplace=True)
```

```python
testO_num=testXO_num.reset_index()
testO_num.drop('index', axis=1, inplace=True)

testT_num=testXT_num.reset_index()
testT_num.drop('index', axis=1, inplace=True)
```

```python
testXA
```

```python
emA=gscv_xgbA.best_estimator_
predA = emA.predict(testXA)

emO=gscv_xgbO.best_estimator_
predO = emO.predict(testXO)

emT=gscv_xgbT.best_estimator_
predT = emT.predict(testXT)
```

```python
predA
```

```python
predA_val=predA.tolist()
predO_val=predO.tolist()
predT_val=predT.tolist()
```

```python
predictedA=pd.DataFrame(predA_val, columns=['Y_Class'])
predictedO=pd.DataFrame(predO_val, columns=['Y_Class'])
predictedT=pd.DataFrame(predT_val, columns=['Y_Class'])
```

```python
finalA=pd.concat([testA_num, predictedA], axis=1)
finalO=pd.concat([testO_num, predictedO], axis=1)
finalT=pd.concat([testT_num, predictedT], axis=1)
```

```python
finalA
```

```python
final_df=pd.concat([finalA, finalO,finalT])
final_df
```
