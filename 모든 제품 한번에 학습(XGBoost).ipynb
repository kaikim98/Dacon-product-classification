```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./train.csv")
data
```

```python
def nan_check(data):
  total=data.isnull().sum().sort_values(ascending=False)
  percent_1 = data.isnull().sum()/data.isnull().count()*100
  percent_2 = (np.round(percent_1, 1)).sort_values(ascending=False)
  missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
  return missing_data

x=nan_check(data)
c_1=x['%'] == 100
li=x[c_1].index.to_list()
li
```

```python
new_li=[]
for i in li:
  i.replace('\n', '')
  new_li.append(i)
new_li
```

```python
for i in li:
  try:
    data.drop(i, axis=1, inplace=True)
  except:
    continue

```

```python
data.drop('TIMESTAMP', axis=1, inplace=True)
```

```python
data
```

```python
trainY=data.iloc[:,1]
trainY
```

```python
new_data=data.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})
```

```python
new_data
```

```python
trainX=new_data.iloc[:, 3:]
trainX
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
param_xgb = {"max_depth": [10,30,50],
              "min_child_weight" : [1,3,6,10],
              "n_estimators": [200,300,500,1000]
              }    
```

```python
gscv_xgb = GridSearchCV (estimator = xgb, param_grid = param_xgb, scoring ='accuracy', cv = 3, refit=True, n_jobs=1, verbose=2)
gscv_xgb.fit(trainX, trainY)
```

```python
print('XGB 파라미터: ', gscv_xgb.best_params_)
print('XGB 예측 정확도: {:.4f}'.format(gscv_xgb.best_score_))
```

```python
data_test = pd.read_csv("./test.csv")
for i in li:
  try:
    data_test.drop(i, axis=1, inplace=True)
  except:
    continue
data_test.drop('TIMESTAMP', axis=1, inplace=True)
new_data_test=data_test.replace({'A_31':0, 'O_31':1, 'T_31':2, 'T050304':0, 'T050307':1, 'T100304':2, 'T100306':3, 'T010306':4, 'T010305':5})
testX=new_data_test.iloc[:,1:]
testX
```

```python
testX
```

```python
from sklearn.metrics import accuracy_score
em=gscv_xgb.best_estimator_
pred = em.predict(testX)
pred
```

```python
pred
```

```python
pred_val=pred.tolist()
```

```python
pred_num=data_test.iloc[:, 0]
pred_num
```

```python
predicted=pd.DataFrame(pred_val, columns=['Y_Class'])
predicted
```

```python
final=pd.concat([pred_num, predicted], axis=1)
final
```

```python
final.to_csv('submission.csv', sep=',')
```
